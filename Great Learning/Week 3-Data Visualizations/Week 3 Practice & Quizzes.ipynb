```python
# Import Dependencies
import numpy as np
import pandas as pd

# Data visualization dependencies
import matplotlib.pyplot as plt
import seaborn as sns

# Command to tell python to display graphs
%matplotlib inline
```

```python
df = pd.read_csv('Placement_Data.csv')
df.head()
```

```python
shape = df.shape
description = df.describe()
information = df.info()
```

```python
print(f'The shape of the dataset is:\n {shape}\n')
print(f'The basic statistical information within the dataset is:')
print(description)
```

```python
placed_candidates = df['status'].value_counts()
placed_candidates
```

```python
sns.histplot(data=df, x='etest_p')
```

```python
sns.countplot(x='hsc_s', data=df, hue='status')
```

```python
sns.countplot(x='hsc_s', data=df)
```

```python
sns.heatmap(data=df[['etest_p','mba_p', 'ssc_p', 'degree_p', 'hsc_p']].corr(), annot=True, cmap ='YlGnBu');
```

```python
sns.boxplot(data=df, x='degree_p', y='hsc_s')
```

```python
sns.lmplot(data=df, x='ssc_p', y='hsc_p', hue='status',ci=False)
```

```python
sns.histplot(data=df, x='degree_p', hue='status', kde=True)
```

```python
not_placed_candidates = df[df['status'] == 'Not Placed']

sns.boxplot(data=not_placed_candidates, x='ssc_p')
```

```python
sns.jointplot(data=df, x='degree_p', y='ssc_p', kind ='hex')
plt.colorbar();
```

```python
sns.histplot(data=df, x='etest_p', kde=True)
```
